버전 1: loop 전부 돌기

````python
# 접근 방법

# 1. 1부터 ~ n까지의 모든 숫자를 검사하여 3의 배수인지 확인
# 2. 3의 배수인 경우 해당 숫자 출력


#문제
def print_multiples_of_three(n):
    for i in range(1, n+1):   # 1~n까지의 범위 지정
        if i % 3 == 0:   # 3의 배수 확인
            print(i)
````

버전 2: 비용 절약, 다 찾으면 바로 종료

````python
# 접근 방법

# 1. 1부터 ~ n까지의 모든 숫자를 검사하여 3의 배수인지 확인
# 2. 3의 배수인 경우 해당 숫자 출력


#문제
def print_multiples_of_three(n):
    for i in range(1, n):
        if (3 * i) <= n: # 다 찾으면, 바로 종료
            print(3 * i)
        else:
            break 
````

````python
#프로그램 실행
print_multiples_of_three(7)
````

- 버전 2: 제거하기, 모음인 것만

````python
# 접근 방법
# 1. 모음에 해당하는 데이터 할당
# 2. 문자열에서 모음 제거 후 문자열 반환

#문제

def remove_vowels(s):
    vowels = "aeiou"
    result = s
    for v in vowels:
        if v in s:
            result=result.replace(v,"")
    return result
````

사용자가 입력한 숫자 n까지의 짝수 합을 구하는 함수를 작성하되, 0을 입력하면 프로그램을 종료하는 함수를 작성하세요. (단, 음수는 입력하지 않는다고 가정합니다.)

- 함수 설명  
`sum_even_numbers()`

  - 사용자에게 숫자 입력을 요청합니다.
  - 0 입력시 종료됨을 안내합니다.
  - 프로그램 종료 시 "1부터 n까지의 짝수 합은 ___ 입니다." 가 출력됩니다.

- 입출력 예시   

    - 입력1:

    ```

    ```

    ```
    "1부터 9까지의 짝수 합은 20입니다."
    ```

    ```
    0
    ```

    - 출력2:

    ```
    (없음)
    ```



- 버전 1: 모든 경우를 다 탐색하는 버전

````python
#문제
def sum_even_numbers():
    while True:
        n = int(input("숫자를 입력하세요 (0 입력 시 종료): "))
        if n == 0:
            break
        total = 0
        for i in range(1, n + 1): #모든 n을 탐색
            if i % 2 == 1: #홀수인 건 pass
                continue
            total += i
        print(f"1부터 {n}까지의 짝수 합은 {total}입니다.")
````

- 버전 2: (비용 절감) 필요한 만큼만 찾는 버전

````python
# 무한 반복 프로그램

### 안내 문구
# 1. "숫자를 입력하세요 (0 입력 시 종료): "
# 2. "1부터 {n}까지의 짝수 합은 {total}입니다."

def sum_even_numbers():
    while True:
        n = int(input("숫자를 입력하세요 (0 입력 시 종료): "))
        # 종료
        if n == 0:
            print("프로그램이 종료되었습니다.")
            break
        i = 1
        sum = 0
        while (2 * i) <= n: # 필요한 짝수만 찾고 종료
            sum += (2 * i)
            i += 1
        print(f"1부터 {n}까지의 짝수 합은 {sum}입니다.")
````

````python
#프로그램 실행
sum_even_numbers()
````

````python
# try-except 구문의 간단한 예시
try:
    # 시도하려는 코드
    number = int(input("Enter a number: "))
    result = 10 / number
    print(f"The result is {result}")
except ValueError:
    # 입력 값이 숫자가 아닐 때 발생하는 예외 처리
    print("Invalid input! Please enter a number.")
except ZeroDivisionError:
    # 0으로 나누려고 할 때 발생하는 예외 처리
    print("You cannot divide by zero!")
````

#### 문제1. 문자열 압축 게임


- 문제 설명  
주어진 문자열에서 연속해서 같은 문자가 반복되는 경우, 그 문자와 반복 횟수를 이용해 새로운 형태로 문자열을 압축합니다. 예를 들어, "aabcccccaaa" 문자열은 "a2b1c5a3"로 압축될 수 있습니다. 압축 결과의 길이가 원본 문자열보다 짧을 경우에만 압축 결과를 반환하고, 그렇지 않으면 원래 문자열을 반환합니다.

- 함수 설명  
`compress_string(s: str) -> str`:  
  - s: 압축하고자 하는 문자열입니다.
  - 압축된 결과가 원본 문자열보다 짧으면 압축된 문자열을 반환하고, 그렇지 않으면 원본 문자열을 반환합니다.

- 입출력 예시   

    - 입력1:

    ```
    "aaabbaaa"
    ```

    - 출력1:

    ```
    "a3b2a3"
    ```

    - 입력2:
    
    ```
    "aabcccccaaa"
    ```

    - 출력2:

    ```
    "a2b1c5a3"
    ```



- 버전 1: for 문 끝나고 한 번 더 입력

````python
# aaabbaaa  = a3b2a3
# s[0] = a , s[1] = a (count = 2)
# s[1] = a , s[2] = a (count = 3)
# 결과 값이 a3

def compress_string(s):
    result = ""
    cnt = 1 #단어 등장 횟수
    i = 0 # 인덱스
    for _ in range(len(s)-1): # len(s) -1 : 뒤의 인덱스를 사용하므로
        if s[i] == s[i+1]: # 앞, 뒤 비교시
           cnt += 1
        else: # 다른 문자 등장시
            result += f"{s[i]}{cnt}"
            cnt = 1 #초기화
        i += 1
    result += f"{s[i]}{cnt}" #마지막 1번 추가
    
    # 압축된 결과가 원본 문자열보다 짧으면 압축된 문자열을 반환하고, 그렇지 않으면 원본 문자열을 반환합니다.
    if len(result) < len(s):
        return result
    else:
        return s
````

- 버전2: for 문 안에서 해결

````python
# aaabbaaa  = a3b2a3
# s[0] = a , s[1] = a (count = 2)
# s[1] = a , s[2] = a (count = 3)
# 결과 값이 a3

def compress_string(s):
    compressed_string = "" # 압축된 결과 저장
    count = 1  # 같은 문자의 반복 수를 카운트

    # 주어진 문자열을 순회하면서 압축 진행
    for i in range(len(s)): 
        # 현재 문자와 다음 문자가 같으면 count를 증가시킴
        if i + 1 < len(s) and s[i] == s[i + 1]: ######### len(s) -1 대신, i + 1 < len(s) 검사
            count += 1
        else:
            # 현재 문자와 다음 문자가 다르거나 문자열의 끝에 도달한 경우
            compressed_string += s[i] + str(count) # aaa -> b = a3
            count = 1  # count 초기화

    # 압축된 문자열이 원래 문자열보다 짧을 경우에만 반환
    if len(compressed_string) < len(s):
        return compressed_string
    else:
        return s
````

````python
# 예시 01
s = "aabcccccaaa"
print(compress_string(s))  # 출력: "a2b1c5a3"
````

````python
# 예시 02
s = "abc"
print(compress_string(s))
````

````python
# @title
import csv
# (시뮬레이션을 위한) 예시 파일 작성코드

def write_tasks_to_csv(filename, tasks):
    with open(filename, 'w', newline='') as csvfile:
        fieldnames = ['할일', '소요시간']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        for task, time in tasks:
            writer.writerow({'할일': task, '소요시간': time})

# 할 일과 소요 시간 데이터
tasks = [
    ('공부하기', 120),
    ('운동하기', 60),
    ('가계부정리', 30),
    ('영화보기', 150)
]

# CSV 파일로 저장
write_tasks_to_csv('tasks.csv', tasks)
````

###### 생각해보기


- 이 문제에서는 `남은 시간을 최적으로 활용하는 방법을 고려`하고 있습니다. 할 일을 선택하는 과정에서 간혹 같은 총 소요 시간 내에서 여러 조합의 할 일 목록을 선택할 수 있는 상황이 발생할 수 있습니다. 이러한 경우, 어떤 기준을 적용하여 할 일의 조합을 결정할까요?      
                  
       

다음과 같은 기준들을 고려해볼 수 있습니다:   
- `가장 많은 할 일을 포함하는 조합`: 가능한 많은 다양한 활동을 하고자 할 때 선택할 수 있는 방법입니다.
    1. `가장 적은 시간이 남는 조합`: 가용 시간을 거의 다 쓰는 조합을 선택하여 시간을 최대한 활용하고자 할 때 선택할 수 있는 방법입니다.
    2. `특정 할 일을 우선하는 조합`: 개인의 선호도나 긴급도에 따라 특정 할 일을 우선적으로 포함시키는 조합을 선택할 수 있습니다.
    3. `가장 많은 시간을 소모하는 할 일을 포함하는 조합`: 가장 긴 시간이 필요한 할 일을 포함시켜, 일상에서 잘 못하던 일을 이번 기회에 해결하고자 할 때 선택할 수 있는 방법입니다.

이러한 선택지 중에서 어떤 기준이 당신의 상황에 가장 적합한지 고민해보고, 이를 코드로 반영해보세요.  

````python
from itertools import combinations

# 예시로 사용할 tasks 리스트

def suggest_optimal_tasks(tasks, remaining_time, criterion="최소_남는시간", priority_task=None):
    """
    남은 시간을 최적 활용하는 기준을 설정하여 할 일을 선택
    :param tasks: 할 일 목록(딕셔너리 리스트)
    :param remaining_time: 남은 시간 (분)
    :param criterion: 최적의 조합을 선택하는 기준 (최소_남는시간, 특정_우선, 최대_소모시간)
    """
    task_combinations = []
    
    # 모든 가능한 할 일 조합을 생성
    for i in range(1, len(tasks) + 1):
        task_combinations += list(combinations(tasks, i))

    optimal_combination = []
    
    if criterion == "최소_남는시간":
        # 1. 가용 시간을 가장 적게 남기는 조합 선택
        min_time_diff = remaining_time # 최솟값 찾기
        for comb in task_combinations:
            total_time = sum(int(task['소요시간']) for task in comb) # 모든 조합의 시간 총합을 계산하기
            if total_time <= remaining_time and (remaining_time - total_time) < min_time_diff:
                min_time_diff = remaining_time - total_time # 모든 조합을 순회하며, 최솟값 탐색
                optimal_combination = comb
                
    elif criterion == "특정_우선":
        # 2. 특정 할 일을 우선 선택 (예: "할 일 3"을 우선)
        # (가정) priority_task = "할 일 3"  # 우선 선택할 할 일 이름
        min_time_diff = remaining_time
        for comb in task_combinations:
            total_time = sum(int(task['소요시간']) for task in comb)
            if any(task['할일'] == priority_task for task in comb) and total_time <= remaining_time: # 특정 할 일이 포함된 조합에 대해서만,
                if (remaining_time - total_time) < min_time_diff: # 최솟값이면,
                    min_time_diff = remaining_time - total_time
                    optimal_combination = comb
    
    elif criterion == "최대_소모시간": # 사실 1번이랑 동일한 결과 -> 단, 최댓값 찾기
        # 3. 가장 많은 시간을 소모하는 조합 선택
        max_total_time = 0
        for comb in task_combinations:
            total_time = sum(int(task['소요시간']) for task in comb) 
            if total_time <= remaining_time and total_time > max_total_time: # 단, 최댓값 찾기
                max_total_time = total_time
                optimal_combination = comb

    # 선택된 최적 조합 출력
    if optimal_combination:
        print(f"선택된 할 일 목록 ({criterion} 기준):")
        for i, task in enumerate(optimal_combination, start=1): # 첫 시작 인덱스 설정 가능!
            print(f"{i}. {task['할일']} - 예상 소요 시간: {task['소요시간']}분")
        remaining = remaining_time - sum(int(task['소요시간']) for task in optimal_combination)
        print(f"위 업무를 모두 수행하고, {remaining} 분이 남습니다. :)")
    else:
        print("주어진 시간 내에 수행 가능한 할 일이 없습니다.")

# 사용 예시
remaining_time = 100  # 남은 시간 (분)
suggest_optimal_tasks(tasks, remaining_time, criterion="최소_남는시간")
print("")
suggest_optimal_tasks(tasks, remaining_time, criterion="특정_우선", priority_task="운동하기")
print("")
suggest_optimal_tasks(tasks, remaining_time, criterion="최대_소모시간")
````

#### 문제2. 수학퍼즐 : 소수의 합

  ```
  10까지의 소수의 합은 17입니다.
  ```

````python
def eratosthenes_sieve(limit):
    # True로 채워진 배열을 만든다. 인덱스가 소수인 경우 True로 남고, 소수가 아닌 경우 False가 된다.
    sieve = [True] * (limit + 1)
    sieve[0] = sieve[1] = False  # 0과 1은 소수가 아니므로 False로 설정
    
    # 2부터 시작하여 각 소수의 배수를 제거
    for i in range(2, int(limit ** 0.5) + 1):
        if sieve[i]:
            # i의 배수들을 False로 표시
            for j in range(i * i, limit + 1, i):
                sieve[j] = False
    
    # True인 인덱스를 소수로 반환
    primes = [i for i, is_prime in enumerate(sieve) if is_prime]
    return primes

def main():
    limit = int(input("어떤 수까지의 소수의 합을 구하시겠습니까?: "))
    primes = eratosthenes_sieve(limit)
    print(f"{limit}까지의 소수의 합은 {sum(primes)}입니다.")

# 프로그램 실행
main()
````

##### `solve_sudoku` 함수

- **문제 설명:**

  이 함수는 주어진 스도쿠 퍼즐을 해결합니다. 주어진 스도쿠 퍼즐이 해결 가능한 경우에만 해결하고, 그렇지 않은 경우에는 메시지를 출력합니다.

- **입력:**

  - 스도쿠 보드: 9x9 크기의 2차원 리스트

- **출력:**

  - 스도쿠 퍼즐이 해결된 경우, 해결된 스도쿠 보드
  - 스도쿠 퍼즐이 해결되지 않은 경우, "해결할 수 없는 스도쿠 퍼즐입니다." 메시지 출력

- **예시:**

  ```python
  board = [
      [5, 3, 0, 0, 7, 0, 0, 0, 0],
      [6, 0, 0, 1, 9, 5, 0, 0, 0],
      [0, 9, 8, 0, 0, 0, 0, 6, 0],
      [8, 0, 0, 0, 6, 0, 0, 0, 3],
      [4, 0, 0, 8, 0, 3, 0, 0, 1],
      [7, 0, 0, 0, 2, 0, 0, 0, 6],
      [0, 6, 0, 0, 0, 0, 2, 8, 0],
      [0, 0, 0, 4, 1, 9, 0, 0, 5],
      [0, 0, 0, 0, 8, 0, 0, 7, 9]
  ]

  print_board(board)
  print("\\n해결된 스도쿠 퍼즐:")
  solved_board = solve_sudoku(board)
  if solved_board:
      print_board(solved_board)
  else:
      print("해결할 수 없는 스도쿠 퍼즐입니다.")

  ```

- **출력 예시:**

  ```
  5 3 0 0 7 0 0 0 0
  6 0 0 1 9 5 0 0 0
  0 9 8 0 0 0 0 6 0
  8 0 0 0 6 0 0 0 3
  4 0 0 8 0 3 0 0 1
  7 0 0 0 2 0 0 0 6
  0 6 0 0 0 0 2 8 0
  0 0 0 4 1 9 0 0 5
  0 0 0 0 8 0 0 7 9

  해결된 스도쿠 퍼즐:
  5 3 4 6 7 8 9 1 2
  6 7 2 1 9 5 3 4 8
  1 9 8 3 4 2 5 6 7
  8 5 9 7 6 1 4 2 3
  4 2 6 8 5 3 7 9 1
  7 1 3 9 2 4 8 5 6
  9 6 1 5 3 7 2 8 4
  2 8 7 4 1 9 6 3 5
  3 4 5 2 8 6 1 7 9

  ```

````python
def is_valid(board, row, col, num):
    # 해당 숫자가 같은 행에 있는지 확인
    if num in board[row]:
        return False
    # 해당 숫자가 같은 열에 있는지 확인
    if num in [board[i][col] for i in range(9)]:
        return False
    # 해당 숫자가 같은 3x3 박스에 있는지 확인
    start_row, start_col = 3 * (row // 3), 3 * (col // 3) # // 몫의 값을 나타낸다. -> 내가 포함된 3x3박스의 좌측 상단 시작점!
    for i in range(start_row, start_row + 3):
        for j in range(start_col, start_col + 3):
            if board[i][j] == num:
                return False
    return True

def find_empty_location(board):
    # 비어 있는 위치(0)를 찾음
    for i in range(9):
        for j in range(9):
            if board[i][j] == 0:
                return i, j
    return None

def solve_sudoku(board):
    # 빈값, 아직 안 푼 곳을 찾고
    empty_loc = find_empty_location(board)
    if not empty_loc: # not None = True
        return True  # 모든 빈칸을 다 채운 경우, 끝
    
    # 빈 위치
    row, col = empty_loc

    for num in range(1, 10): # 빈값에 1~9를 다 넣어보면서, 유효한 값을 찾는다.
        if is_valid(board, row, col, num):
            board[row][col] = num # 유효하면, 대입!

            # 위 규칙에 의거, 값을 채워 넣었는데, 결국 스도쿠를 해결하지 못했다면,
            if solve_sudoku(board):
                return True

            board[row][col] = 0  # 실패한 경우, 백트래킹

    return False  # 이 경로로는 해결 불가

# 주어진 스도쿠 퍼즐을 해결하는 함수 호출
board = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

print("초기 스도쿠 퍼즐:")
print_board(board)
print("\n해결된 스도쿠 퍼즐:")

if solve_sudoku(board):
    print_board(board)
else:
    print("해결할 수 없는 스도쿠 퍼즐입니다.")
````

##### `generate_sudoku`함수



- **문제 설명:**

  이 함수는 빈 칸이 채워지지 않은 완성된 스도쿠 퍼즐을 생성합니다.

- **입력:**

  - 없음

- **출력:**

  - 완성된 스도쿠 퍼즐을 나타내는 9x9 크기의 2차원 리스트

- **예시:**

  ```python
  sudoku_puzzle = generate_sudoku()
  print_board(sudoku_puzzle)

  ```

- **출력 예시:**

  ```
  5 3 0 0 7 0 0 0 0
  6 0 0 1 9 5 0 0 0
  0 9 8 0 0 0 0 6 0
  8 0 0 0 6 0 0 0 3
  4 0 0 8 0 3 0 0 1
  7 0 0 0 2 0 0 0 6
  0 6 0 0 0 0 2 8 0
  0 0 0 4 1 9 0 0 5
  0 0 0 0 8 0 0 7 9

  ```

````python
import random

def generate_sudoku():
    
    puzzle = []
    # 0의 확률을 50%으로 고정 - 전체 18개 중 9개가 0이므로!
    numbers = [0] * 9 + list(range(1, 10))
    for _ in range(9):    
        row = []
        for _ in range(9):
            #rand_int = random.randint(0, 9)
            rand_int = random.choice(numbers)
            row.append(rand_int)
        puzzle.append(row)
        
    return puzzle

sudoku_puzzle = generate_sudoku()
print_board(sudoku_puzzle)
````

##### `play_sudoku`함수

- **문제 설명:**

  이 함수는 사용자에게 스도쿠 게임을 플레이할 기회를 제공합니다. 게임은 사용자가 잘못된 이동을 하거나 스도쿠를 완성할 때까지 진행됩니다.

- **입력:**

  - 없음. 사용자는 콘솔을 통해 행, 열, 숫자를 입력합니다.

- **출력:**

  - 없음. 스도쿠 보드의 상태가 계속 업데이트되며, 최종적으로 스도쿠가 완료되었거나 잘못된 이동이 있을 때 게임이 종료됩니다.

- **예시:**

  ```python
  play_sudoku()

  ```

- **출력 예시:**

  ```diff

  - - - - - 5 - - 9

  5 - - - - 8 - - -

  숫자를 입력하세요 (1-9): 2

  - - - - - - 9 - -

  - - 6 - 1 - 7 - -
  - - - 4 6 - - - -

  열을 입력하세요 (1-9): 2
  숫자를 입력하세요 (1-9): 5
  잘못된 이동입니다. 게임을 종료합니다.

  ```

````python
import random

# 사용자가 직접 스도쿠를 플레이하는 함수
def play_sudoku():
    board = generate_sudoku()  # 스도쿠 퍼즐을 생성
    print("스도쿠 게임을 시작합니다!")
    print_board(board)

    while True:
        # 사용자가 입력한 값으로 행, 열, 숫자를 받음
        try:
            row = int(input("행을 입력하세요 (1-9): ")) - 1
            col = int(input("열을 입력하세요 (1-9): ")) - 1
            num = int(input("숫자를 입력하세요 (1-9): "))

            # 유효한 입력인지 확인
            if not (0 <= row <= 8 and 0 <= col <= 8 and 1 <= num <= 9):
                print("잘못된 입력입니다. 1에서 9 사이의 숫자를 입력하세요.")
                continue

            if board[row][col] != 0:
                print("이미 값이 입력된 칸입니다.")
                continue

            # 숫자가 해당 위치에 올바르게 입력될 수 있는지 확인 (검증)
            if is_valid(board, row, col, num):
                board[row][col] = num  # 입력한 숫자를 보드에 반영
                print_board(board)

                # 스도쿠가 완성되었는지 확인
                if not find_empty_location(board): # 빈 곳을 못 찾으면!
                    print("축하합니다! 스도쿠를 완성했습니다.")
                    break
            else:
                print("잘못된 이동입니다. 게임을 종료합니다.")
                break
        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력하세요.")
        
# 프로그램 실행
play_sudoku()
````
